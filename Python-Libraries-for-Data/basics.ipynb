```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

plt.plot(x, y, marker='o', linestyle='-', color='green')
plt.title('Line chart example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
cat = ['cat a', 'cat b', 'cat c', 'cat d']
values = [25, 18, 30, 22]
plt.title('bar chart example')
plt.bar(cat, values, color='green')
plt.xlabel('values') 
plt.show()
```

```python
data = [1,2,2,3,3,3,4,4,4,5,5,5,5,5]
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('histogram example')
plt.xlabel('values')
plt.ylabel('frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

sizes = [30, 35, 20, 15, 10]
cat = ['cat a', 'cat b', 'cat c', 'cat d', 'cat e']
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'orange']

plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=cat, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Pie Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]

plt.figure(figsize=(4, 4))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Pie Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

group1 = [10, 15, 20, 25, 30]
group2 = [5, 10, 15, 20, 25]
group3 = [15, 18, 22, 28, 35]

plt.boxplot([group1, group2, group3], labels=['group1', 'group2', 'group3'])
plt.title('Box Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.show()
```
